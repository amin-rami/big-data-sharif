```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('MOVIE_IMDB.csv')
```

```python
print('The number NaN values of each row:')
df.apply(lambda row: row.isna().sum(), axis=1)
```

```python
print('The number NaN values of each column:')
df.isna().sum()
```

```python
print('Percentage of NaN values in each column:')
round(df.isna().sum()/df.shape[0]*100, 2)
```

```python
columns_to_delete = [
    'color',
    'director_facebook_likes',
    'actor_1_facebook_likes',
    'actor_2_facebook_likes',
    'actor_3_facebook_likes',
    'actor_2_name',
    'cast_total_facebook_likes',
    'actor_3_name',
    'duration',
    'facenumber_in_poster',
    'content_rating',
    'country',
    'movie_imdb_link',
    'aspect_ratio',
    'plot_keywords',
]

df.drop(columns_to_delete, axis=1, inplace=True)
df
```

```python
print('Percentage of NaN values in each column:')
round(df.isna().sum()/df.shape[0]*100, 2)
```

```python
df.dropna(subset=['gross', 'budget'], inplace=True)
```

```python
df['language'] = df['language'].fillna('English')
```

```python
print('Percentage of NaN values in each column:')
round(df.isna().sum()/df.shape[0]*100, 2)
df
```

```python
print('dropping duplicates')
df.drop_duplicates(inplace=True)
df
```

```python
df['profit'] = df['gross'] - df['budget']
df
```

```python
df_positive_profit = df[df['profit'] >= 0]
df_negative_profit = df[df['profit'] < 0]

fig = plt.figure(figsize=(12, 3))
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(df_positive_profit['budget'], df_positive_profit['profit'], 'bo')
ax1.set_xlabel('budget')
ax1.set_ylabel('profit')
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_title('budget vs profit plot for movies with positive profit')
ax1.grid()


fig = plt.figure(figsize=(12, 3))
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(df_negative_profit['budget'], df_negative_profit['profit'], 'bo')
ax2.set_xlabel('budget')
ax2.set_ylabel('profit')
ax2.set_xlim([0, 2e8])
ax2.set_ylim([-2e8, 0])
ax2.set_title('budget vs profit plot for movies with negative profit')
ax2.grid()
```

```python
print('Movies with top profit')
df.sort_values('profit', ascending=False).head(10).reset_index()[['movie_title', 'profit']]
```
